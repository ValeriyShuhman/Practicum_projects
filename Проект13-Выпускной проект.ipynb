Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. Ваша задача — построить модель, которая будет её предсказывать.

# Описание этапа обработки.

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

Составим список ковшей, для которых в датасете нет начальной или конечной температуры, чтобы убрать их из датасета позже.

Составим список ковшей, в которые добавляли сыпучие материалы(присадки) и проволочные материалы(присадки) до первого измерения температуры, чтобы убрать эти ковши из датасета позже, т.к. если процесс начинается не с измерения температуры, то такие ковши не качественные.

## Объединение данных.

**Объединим выбранные нами данные в один датафрейм для последующего анализа.**

Создадим несколько признаков для улучшения качества моделирования.

```python
#Создадим таблицу с признаком 'полная_мощность'
data_ac_power = data_arc_new[['key', 'активная_мощность', 'реактивная_мощность']]
data_ac_power['полная_мощность'] = (data_arc_new['активная_мощность']**2 + data_arc_new['реактивная_мощность']**2)**0.5
data_full_power = pd.pivot_table(data_ac_power, values=['полная_мощность'], index='key', aggfunc='sum')
data_full_power = data_full_power.reset_index()

#Создадим таблицу с признаком 'начальная_температура'
data_start_temp_time = pd.pivot_table(data_temp_new, values=['время_замера'], index='key', aggfunc='min')
data_start_temp = data_start_temp_time.merge(data_temp_new, on='время_замера')
data_start_temp = data_start_temp.drop('время_замера', axis=1)
data_start_temp.columns = ['key', 'начальная_температура']

#Создадим таблицу с признаком 'конечная_температура'
data_end_temp_time = pd.pivot_table(data_temp_new, values=['время_замера'], index='key', aggfunc='max')
data_end_temp = data_end_temp_time.merge(data_temp_new, on='время_замера')
data_end_temp = data_end_temp.drop('время_замера', axis=1)
data_end_temp.columns = ['key', 'конечная_температура']

#Создадим таблицу с признаком 'длительность_работы_электродов'
data_duration_heating = data_arc_new.copy()
data_duration_heating['duration'] = data_duration_heating['конец_нагрева_дугой'] - data_duration_heating['начало_нагрева_дугой']
data_duration_heating = pd.pivot_table(data_duration_heating, values=['duration'], index='key', aggfunc='sum')
data_duration_heating = data_duration_heating.astype('int')
data_duration_heating = data_duration_heating / 3600000000000
data_duration_heating.columns = ['длительность_работы_электродов']
data_duration_heating = data_duration_heating.reset_index()
```

## Обучение моделей машинного обучения.

**Модель LinearRegression.**

```python
def my_mae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

mae_scorer = make_scorer(my_mae, greater_is_better=False)
```

```python
%%time

# Обучим модель LinearRegression
m_linreg = LinearRegression()

m_linreg.fit(features_train, target_train)

#Определяю MAE модели с помощью кросс-валидации
regression_cv_score = pd.Series(cross_val_score(m_linreg,
                                                features_train,
                                                target_train,
                                                scoring=mae_scorer,
                                                cv=5)).mean()*(-1)

print('Среднее качество модели линейной регрессии на кросс-валидации:', regression_cv_score)
```

**Модель RandomForestRegressor.**

```python
%%time

# Обучим модель RandomForestRegressor()
rfr = RandomForestRegressor()

# Задаём параметры
rfr_param = { 'n_estimators': range(5, 25, 5),
                     'max_depth': range(1, 20, 2),
                     'min_samples_leaf': range(1, 7),
                     'min_samples_split': range(2, 6, 2) }

rfr_grid = GridSearchCV(rfr,
                        rfr_param, 
                        scoring=mae_scorer, 
                        cv=5)

rfr_grid.fit(features_train, target_train)
```

```python
print('Среднее качество модели RandomForestRegressor на кросс-валидации:', rfr_grid.best_score_*(-1))
```

```python
%%time

# Обучим модель CatBoostRegressor()
cbr = CatBoostRegressor()

# Задаём параметры
cbr_params = {'learning_rate': np.logspace(-3, 0, 5),
              'iterations': range(20, 80, 20),
              'depth': range(5,10,2) }

cbr_grid = GridSearchCV(cbr, 
                       cbr_params, 
                       scoring=mae_scorer, 
                       cv=5)

cbr_grid.fit(features_train, target_train, verbose=0)
```

```python
print('Среднее качество модели CatBoostRegressor на кросс-валидации:', cbr_grid.best_score_*(-1))
```
