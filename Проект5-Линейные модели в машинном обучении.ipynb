**Проект: Линейные модели в машинном обучении.**

**Описание проекта**

Вы работаете в IT-компании, которая выполняет на заказ проекты по машинному обучению. К вам обратился фермер, владелец молочного хозяйства «Вольный луг». Он хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».

Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. А продавцы и технологи так и норовят приукрасить своих коровок!

Поэтому он просит вас разработать модель машинного обучения, которая поможет ему управлять рисками и принимать объективное решение о покупке. «ЭкоФерма» готова предоставить подробные данные о своих коровах. Вам нужно создать две прогнозные модели для отбора бурёнок в поголовье:

- Первая будет прогнозировать возможный удой коровы (целевой признак Удой);
- Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).

С помощью модели нужно отобрать коров по двум критериям:

- средний удой за год — не менее 6000 килограммов;
- молоко должно быть вкусным.

Признаки в датасете ferma_main.csv можно разделить на группы.

**Подготовим данные датасета ferma_main.csv и обучим на них первую модель LinearRegression.**

```python
#Разделим данные на целевой и входные признаки
RANDOM_STATE = 42

X = data_fm.drop('удой, кг', axis=1)
y = data_fm['удой, кг']
```

Если делать выводы опираясь на причинно-следственные связи между удоем коров и группами признаков из описания датасета, то для количественных признаков лучше всего выбрать: эке (энергетическая кормовая единица); сырой протеин, г; спо (сахаро-протеиновое соотношение), так как они являются предпосылками для удойности коровы. Из тех же соображений для категориальных признаков лучше всего выбрать признаки: порода; тип пастбища; порода папы_быка и возраст. Мы не будем брать в модель признаки свойств молока, такие как жирность, так как с изменением питания они могут измениться.

```python
#Отберем признаки, которые подходят для модели линейной регрессии
num_data = ['эке (энергетическая кормовая единица)', 'сырой протеин, г', 'спо (сахаро-протеиновое соотношение)']
cat_data = ['порода', 'тип пастбища', 'порода папы_быка', 'возраст']
```

```python
#Разобьем данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE)
```

```python
#Кодируем категориальные признаки с помощью OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[cat_data])
X_test_ohe = encoder.transform(X_test[cat_data])
```

```python
#Масштабируем количественные признаки с помощью StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_data])
X_test_scaled = scaler.transform(X_test[num_data])
```

```python
#Создадим датафреймы для объединения
X_train_ohe = pd.DataFrame(X_train_ohe, columns=cat_data)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=cat_data)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_data)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_data)
#Объединим их методом concat
X_train_ohe_sc = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test_ohe_sc = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
#Обучим модель линейной регрессии на подготовленных данных
model_lr = LinearRegression()
model_lr.fit(X_train_ohe_sc, y_train)
predictions_1 = model_lr.predict(X_test_ohe_sc)
```

```python
#Рассчитаем R²
r_2 = r2_score(y_test, predictions_1)  
r_2
```

```python
#Оценим качество модели на тестовой выборке, проанализируем остатки с помощью графиков.
residuals = y_test - predictions_1
#Построим гистограмму распределения остатков
plt.hist(residuals, bins=50)
plt.title('Распределение остатков')
plt.xlabel('Распределение остатков')
plt.show()
```

```python
plt.scatter(predictions_1, residuals)
plt.title('Распределение остатков')
plt.xlabel('Предсказания удоев')
plt.ylabel('Распределение остатков')
plt.show()
```

**Вывод:** Коэффициент детерминации R2 равен 0.784 для данной модели, это среднее значение.
Анализ остатков дал симметричное нормальное распределение. Это значит, что среднее значение остатков модели равно 0. Такая модель работает качественно. Распределение остатков равномерное.

```python
#Отберем признаки, которые подходят для модели линейной регрессии
num_data = ['эке (энергетическая кормовая единица)', 'сырой протеин, г', 'жирность,%', 'белок,%']
cat_data = ['порода', 'тип пастбища', 'порода папы_быка', 'возраст', 'спо (сахаро-протеиновое соотношение)', 'имя папы']
```
